### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene":{"$eq":20}}
results = establishments.find(query)

# Use count_documents to display the number of documents in the result
print(f"No of documents in the result is: {establishments.count_documents(query)}")
# Display the first document in the results using pprint
print("Result of the first document:")
pprint(results[0], sort_dicts=False)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f'No of rows in the DataFrame are: {len(df)}')
# Display the first 10 rows of the DataFrame
df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    "RatingValue":{"$gte":4},
    "LocalAuthorityName":{"$regex":"London"}
    }
results = establishments.find(query)

# Use count_documents to display the number of documents in the result

print(f"No of documents in the result is: {establishments.count_documents(query)}")

# Display the first document in the results using pprint
print("Result of the first document:")
pprint(results[0], sort_dicts=False)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f'No of rows in the DataFrame are: {len(df)}')
# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = establishments.find({"BusinessName":"Penang Flavours"})[0]['geocode']['latitude']        #latitude of restaurant
longitude = establishments.find({"BusinessName":"Penang Flavours"})[0]['geocode']['longitude']      #longitude of restaurant

query = {                                                               ### FILTERS ###
    "RatingValue":{"$eq":5},                                            #   RatingValue = 5
    "geocode.latitude":{"$lte":latitude + degree_search},               #   latitude less than max degree search
    "geocode.latitude":{"$gte":latitude - degree_search},               #   latitude greater than max degree search
    "geocode.longitude":{"$lte":longitude + degree_search},             #   longitude less than max degree search
    "geocode.longitude":{"$gte":longitude - degree_search},             #   longitude greater than max degree search
    "scores.Hygiene":{"$gte":0}                                         #   Hygiene score greater than or equal to 0, this
    }                                                                   #   eliminates the "None" values in results for Hygiene

sort = [("scores.Hygiene",1)]                                           # sort by Hygiene score - ascending


limit = 5                                                               # limit to first 5 results

results = establishments.find(query).sort(sort).limit(limit)

# Use count_documents to display the number of documents in the query

print(f"No of documents in the query (when results are not limited) is: {establishments.count_documents(query)}")


# Print the results
pprint(list(results), sort_dicts=False)

```

```python
results = establishments.find(query).sort(sort).limit(limit)

# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f'No of rows in the DataFrame are: {len(df)}')
# Display the first 10 rows of the DataFrame
df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0

match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': "$LocalAuthorityName", 
                          'count': { '$sum': 1 }}}#,
                          #'max_height': { '$max': '$measurements.elementMeasurements.Height' }}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': { 'count': -1} }#, '_id': 1 }}

# Print the number of documents in the result
pipeline = [match_query, group_query, sort_values]
results = list(establishments.aggregate(pipeline))

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(f'No of rows in the DataFrame are: {len(df)}')
# Display the first 10 rows of the DataFrame
df.head(10)
```
